```python
import sys
import os
import numpy as np
import torch
import glob
import matplotlib.pyplot as plt

sys.path.insert(1, "/home/maxim/Projects/ALF/ALF/") 
sys.path.insert(1,"/home/maxim/Projects/TESTS/ALF/PythonPath/") # Paths to ASE and HIPNN
os.environ["HIPPYNN_USE_CUSTOM_KERNELS"] = "pytorch"

from ase import units

import hippynn
from ase.io import read
from ase import units
from ase.md.verlet import VelocityVerlet
from ase.md.langevin import Langevin
import alframework
from matplotlib.pyplot import figure

def HIPNN_ASE_calculator(HIPNN_model_directory,energy_key='energy',device="cuda:0"):
    from hippynn.experiment.serialization import load_checkpoint_from_cwd
    from hippynn.tools import active_directory
    from hippynn.interfaces.ase_interface import HippynnCalculator
    import torch
    with active_directory(HIPNN_model_directory):
        bundle = load_checkpoint_from_cwd(map_location="cpu", restore_db=False)
    model = bundle["training_modules"].model
    energy_node = model.node_from_name(energy_key)
    calc = HippynnCalculator(energy_node, en_unit=units.eV)
    calc.to(torch.float32)
    calc.to(torch.device(device))
    return(calc)
    
def HIPNN_ASE_load_ensemble(HIPNN_ensemble_directory,device="cuda:0"):
    model_list = []
    for cur_dir in glob.glob(HIPNN_ensemble_directory + '/model-*/'):
        model_list.append(HIPNN_ASE_calculator(cur_dir,device=device))
    return(model_list)

from alframework.samplers.ASE_ensemble_constructor import MLMD_calculator
```

```python
# Regular Langevin MD

hipnn_ens = HIPNN_ASE_load_ensemble("/home/maxim/Projects/ALF/model-0098/") # Path to HIPNN ensemble
mlmd = MLMD_calculator(hipnn_ens)
atoms = read("/home/maxim/Projects/ALF/ALF/examples/UDD/fragment_library/glyc_opt.xyz")
atoms.calc = mlmd
#atoms.get_potential_energy()

global std_reg, epot_reg
std_reg = np.array([])
epot_reg = np.array([])

#dyn = VelocityVerlet(atoms, timestep=1.0*units.fs, trajectory=None, loginterval=1, append_trajectory=False)
dyn = Langevin(atoms, timestep=0.8*units.fs, temperature=300*units.kB, friction=0.01, trajectory=None, loginterval=1)
def bias_to_dyn(d=dyn, atoms=atoms):  # store a reference to atoms in the definition.
    global std_reg, epot_reg
    """Function to print the potential, kinetic and total energy."""
    print('N: ', dyn.get_number_of_steps())
    print('Epot: ', atoms.get_potential_energy(), 'Ekin:', atoms.get_kinetic_energy())
    mlmd.calculate(atoms, ['energy_stdev'])
    print('energy_stdev: ',mlmd.results['energy_stdev'],'\n')
    std_reg = np.append(std_reg, mlmd.results['energy_stdev'])
    epot_reg = np.append(epot_reg, atoms.get_potential_energy())
    #print('E_en_bias: ',mlmd.results['E_en_bias'])
dyn.attach(bias_to_dyn, interval=1)
dyn.run(300 )
```

```python
# Regular Langevin MD + UDD

hipnn_ens = HIPNN_ASE_load_ensemble("/home/maxim/Projects/ALF/model-0098/")
mlmd = MLMD_calculator(hipnn_ens)
atoms_udd = read("/home/maxim/Projects/ALF/ALF/examples/UDD/fragment_library/glyc_pert.xyz")
atoms_udd.calc = mlmd

mlmd.use_bias = True # enable Energy std bias
mlmd.E_en_bias_weight = 0.45 # bias magnitude.

global std_bias, epot_bias
std_bias = np.array([])
epot_bias = np.array([])

#dyn = VelocityVerlet(atoms_udd, timestep=1.0*units.fs, trajectory=None, loginterval=1, append_trajectory=False)
dyn = Langevin(atoms_udd, timestep=0.8*units.fs, temperature=300*units.kB, friction=0.01, trajectory=None, loginterval=1)

def bias_to_dyn(d=dyn, atoms=atoms_udd):  # store a reference to atoms in the definition.
    global std_bias, epot_bias
    print('N: ', dyn.get_number_of_steps())
    print('Epot: ', atoms.get_potential_energy() - mlmd.results['E_en_bias'],'Ekin:', atoms.get_kinetic_energy())
    print('energy_stdev: ',mlmd.results['energy_stdev'])
    print('E_en_bias: ',mlmd.results['E_en_bias'], '\n')
    std_bias = np.append(std_bias, mlmd.results['energy_stdev'])
    epot_bias = np.append(epot_bias, atoms.get_potential_energy() - mlmd.results['E_en_bias'])
dyn.attach(bias_to_dyn, interval=1)
dyn.run(300 )
```

```python
plt.rcParams["figure.figsize"] = (12,5)
```

```python
plt.plot(std_reg, label='Langevin')
plt.plot(std_bias, label='Langevin + UDD')
plt.xlabel('MD step')
plt.ylabel('std(E)')
plt.title("Energy STD")
plt.legend()
```

```python
plt.plot(epot_reg, label='Langevin')
plt.plot(epot_bias, label='Langevin + UDD')
plt.xlabel('MD step')
plt.ylabel('E pot')
plt.title("Potential Energy")
plt.legend()
```
